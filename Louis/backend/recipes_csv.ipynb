```python

#? Extract directions for a name of recipe

def get_recipe(title):
    return df_recipes[df_recipes["Title"] == title].Directions.values[0]

print(get_recipe("Crazy Cake"))
```

> ### Test - Extract ingredients

```python

#? Extract ingredients for a name of recipe
def get_ingredients(title):
    list_ingredients = []
    for i in range(1, 19):
        number = str(i).zfill(2)
        ingredient = df_recipes[df_recipes["Title"] == title][f"Ingredient{number}"].values[0]
        ingredient = str(ingredient)
        if ingredient != "nan" and ingredient != "None":
            list_ingredients.append(ingredient)
    return list_ingredients

my_list = get_ingredients("Crazy Cake")
print(my_list)
```
